```python
import pymysql

def show_page(page_number=1, per_page=10):
    try:
        connection = pymysql.connect(**MYSQL_CONFIG)
    
        with connection.cursor() as cursor:
            offset = (page_number - 1)*per_page
    
            query = 'SELECT film.title,  film.release_year, film.rating, film.length FROM film LIMIT %s OFFSET %s'
            cursor.execute(query,(per_page, offset))
            results = cursor.fetchall()
    
            for i, row in enumerate(results, 1):
                print(f'{i}. {row}')
            print('_'*60)
    except:
        print(' ')
    finally:
        cursor.close()
        connection.close()
```

### Арифметические операции над строками:

Строки в Python поддерживают операции сложения и умножения. Операция сложения позволяет объединять строки, а операция умножения позволяет повторять строку определенное количество раз. Например:


```python
greeting = "Hello, "
name = "John"
message = greeting + name  # Конкатенация строк
multiplied_string = name * 3  # Повторение строки
print(message)
print(multiplied_string)
```

Строки ***неизменяемы***, но если вы хотите изменить строку:

```python
s = "Hello"
s = s + " world!" # s+=" world!"
print(s)
```

### Кодировка
Кодировка — это способ преобразования символов в числа, которые компьютер может хранить и обрабатывать. Кодировка позволяет компьютерам работать с текстами на разных языках и различными символами.

#### Кодировка ASCII
ASCII (American Standard Code for Information Interchange) — это одна из первых и простейших кодировок, разработанная для представления текстовых символов в компьютерах. Она определяет стандартные коды для 128 символов, включая буквы английского алфавита, цифры, символы и управляющие коды.

#### Основные характеристики кодировки ASCII:
1. **7 бит:**
* ASCII использует 7 бит для кодирования символов.
* Это позволяет закодировать 128 символов (2^7 = 128).
* Символы представлены числами от 0 до 127.
2. **Символы:**
* 0-31: Управляющие символы (например, символ новой строки, возврат каретки, сигнал окончания текста).
* 32-126: Печатные символы, включая буквы, цифры, пунктуацию и другие символы.
3. **Простота и совместимость:**
* ASCII был создан для работы с английским алфавитом, цифрами и основными символами.
* Он широко используется в компьютерах, сетевых системах и других устройствах, обеспечивая базовый уровень кодирования для текстовых данных.


#### Расширения ASCII
Для языков с нелатинской письменностью (кириллической, греческой, арабской и др.) на основе оригинального ASCII были разработаны 8-битные расширения. Они позволяют кодировать 256 различных символов, из которых первая половина кодовой таблицы (от 0 до 127 позиции) совпадает с ASCII, а вторая служит для размещения национальных и других дополнительных символов. Например, расширение cp-1251 для криллицы выглядит так:


К сожалению, отсутствие единого подхода к размещению национальных символов в их кодовых таблицах (например, кириллицы в КОИ-8, Windows-1251, ISO 8859-5 и т. д.) порождало множество проблем. Так, например, когда почтовые клиенты отправителя и получателя электронного письма использовали различные кодировки для его интерпретации, текст письма (за исключением символов ASCII) становился нечитаемым.

#### Таблица символов Unicode
Unicode — это глобальный стандарт, который присваивает каждому символу уникальный числовой код, называемый кодовой точкой. Он охватывает символы всех языков мира, математические символы, эмодзи и многое другое. Unicode решает проблему ограниченности старых кодировок, таких как ASCII, и позволяет представлять миллионы различных символов.
На данный момент в Unicode определено более 143 000 символов. Это число постоянно растёт, поскольку стандарт Unicode обновляется и добавляются новые символы для различных языков, символов, технических знаков и эмодзи.

#### Категории символов в Unicode включают:

#### Кодировка UTF-8
UTF-8 (Unicode Transformation Format - 8-bit) — это одна из самых популярных и широко используемых кодировок для представления символов в Unicode. Она преобразует символы Unicode в последовательности байтов переменной длины, где каждый символ может занимать от 1 до 4 байт. UTF-8 стала стандартом для передачи и хранения текстов в Интернете благодаря её гибкости и экономичности.


Unicode — это абстрактная таблица символов, а UTF-8 — способ компактно записать эти символы в байтах.

#### Преимущества UTF-8:
1. Совместимость:
* UTF-8 поддерживает полную совместимость с ASCII, что делает его удобным для интеграции с системами, которые изначально использовали только ASCII.
2. Эффективное использование памяти:
* Для текстов на латинице (английский, испанский и т.д.) UTF-8 экономит память, используя только 1 байт на символ.
* Для текстов на других языках, таких как русский или китайский, UTF-8 остаётся эффективной, занимая меньше памяти по сравнению с UTF-16 или UTF-32.
3. Гибкость:
* UTF-8 может кодировать символы любой сложности: от простых букв латинского алфавита до редких символов и эмодзи.
4. Международный стандарт:
* UTF-8 стал основным стандартом кодирования текстов в Интернете. Большинство веб-страниц и веб-приложений используют UTF-8 для работы с текстом на разных языках.


### Функции ord(), chr()

Функция ord() возвращает числовое представление символа, а функция chr() выполняет обратное преобразование - возвращает символ по его числовому представлению. Например:


```python
code = ord('A')  # Числовое представление символа 'A'
character = chr(65)  # Символ по числовому представлению 65
print(code)
print(character)
```

```python
print(ord('B'))
```

```python
#Преобразование строки в байты (str → bytes)
text = "Привет, мир!"
bytes_data = text.encode(encoding="utf-8")  # Можно просто .encode() — UTF-8 по умолчанию
print(bytes_data) 
```

```python
# Преобразование байтов в строку
decoded_text = bytes_data.decode("utf-8")  # Указываем ту же кодировку
print(decoded_text)
```

### Сравнение строк. Понятие лексикографического сравнения

В Python строки могут быть сравнены с помощью операторов сравнения (==, !=, <, >, <=, >=). При сравнении строк применяется лексикографическое сравнение, что означает, что строки сравниваются посимвольно, сравнивая их числовые значения кодов символов в таблице символов. Например:

```python
str1 = "apple"
str2 = "banana"
result = str1 < str2  # Сравнение строк по лексикографическому порядку
print(result)
```

```python
str1 = "apple"
str2 = "b"
result = str1 < str2  # Сравнение строк по лексикографическому порядку
print(result)
```

### Доступ к элементу по индексу. Отрицательные индексы

Строки в Python являются последовательностью символов, и доступ к отдельным символам осуществляется по индексу. Положительные индексы начинаются с 0 и указывают на позицию символа в строке, а отрицательные индексы считаются с конца строки, где -1 соответствует последнему символу. Например:

```python
text = "Hello"
print(text[len(text)-1])
```

```python
text = "Hello"
first_char = text[0]  # Получение первого символа
last_char = text[-1]  # Получение последнего символа
print(first_char)
print(last_char)
```

```python
text = "Hello world!!"
if len(text)%2 == 0:
    print('Середина не выпадает на символ')
else:
    print(text[len(text)//2])
    
```

### Неизменяемость строк, демонстрация этого с помощью id()

Строки в Python являются неизменяемыми объектами, что означает, что после создания их содержимое не может быть изменено. Демонстрация этого может быть выполнена с помощью функции id(), которая возвращает идентификатор объекта. Например:

```python
text = "Hello"
text_id = id(text)  # Получение идентификатора объекта
print(text_id)
```

```python
str1 = 'Привет'
str1[0] = 'п'
print(str1)
```

```python
original_string = "Hello, World!"
print(original_string, id(original_string))

new_string = original_string.replace("H", "h")
print(new_string, id(new_string))

# Проверяем, изменился ли идентификатор оригинальной строки
print("Оригинальная строка после изменения:", original_string, "id:", id(original_string))
```

### Взятие среза от строки. Варианты с 1, 2 и 3 аргументами

В Python для извлечения подстроки из строки используется операция взятия среза. С помощью срезов можно выбрать определенный диапазон символов. Варианты срезов включают использование одного, двух или трех аргументов. Один аргумент указывает индекс символа, с которого начинается срез. Два аргумента указывают начальный и конечный индексы среза. Три аргумента позволяют указать шаг, с которым будут выбраны символы. Например:

```python
text = "Hello, world!"
substring1 = text[0:5]  # Взятие среза с 1-го по 5-й символы
substring2 = text[7:]  # Взятие среза с 8-го символа до конца строки
substring3 = text[::2]  # Взятие среза с шагом 2
print(substring1)
print(substring2)
print(substring3)
```

### Создание копии строки. Разворот строки с помощью срезов:

С помощью срезов можно создавать копии строк. Также можно использовать срезы с отрицательным шагом для разворота строки. Например:


```python
text = "Hello"
copy = text[:]  # Создание копии строки
reversed_text = text[::-1]  # Разворот строки
print(reversed_text)
```

### Метод in для проверки вхождения подстроки в строку

Метод in в Python позволяет проверить, содержится ли одна строка в другой. Он возвращает значение True, если подстрока найдена, и False, если подстрока отсутствует. Например:

```python
text = "Hello, world!"
is_found = "world" in text  # Проверка наличия подстроки
print(is_found)
```

```python
text = "Hello, world!"
is_found = "car" in text  # Проверка наличия подстроки
print(is_found)
```

```python
text = "Hello, world!"
is_found = "!" in text  # Проверка наличия подстроки
print(is_found)
```

```python
'h' in "Hello, world!"
```
